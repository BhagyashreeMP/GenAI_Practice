```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://www.hcltech.com/careers/lead-developertech-lead")
page_data = loader.load().pop().page_content
print(page_data)
```

```python
from langchain_core.prompts import PromptTemplate
prompt_extract = PromptTemplate.from_template(
    """
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the career's page of a website.
        Your job is to extract the job postings and return them in JSON format containing the
        following keys: `role`, `experience`, `skills` and `description`.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):
        """
)

chain_extract = prompt_extract | llm
res = chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser
json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
if isinstance(json_res,list) and len(json_res)>0:
  json_res = json_res[0]

print(json_res)
print(json_res['skills'])
```

```python
import pandas as pd
df = pd.read_csv("my_portfolio.csv")
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="portfolio")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["Techstack"],
                       metadatas={"links": row["Links"]},
                       ids=[str(uuid.uuid4())])
```

```python
job = json_res
links = collection.query(query_texts=job['skills'],n_results =2).get('metadatas',[])
links
```

```python
prompt_email = PromptTemplate.from_template(
        """
        ### JOB DESCRIPTION:
        {job_description}

        ### INSTRUCTION:
        You are Bhagyashree, a Technical Lead at HCL Technologies. Bhagyashree has 2 years experience as Full stack developer.
        Over her experience, she have empowered numerous client with tailored solutions, fostering scalability,
        process optimization, cost reduction, and heightened overall efficiency.
        Your job is to write a cold email to the client regarding the job mentioned above describing the capability of Bhagyashree
        in fulfilling their needs.
        Also add the most relevant ones from the following links to showcase Bhagyashree's portfolio: {link_list}
        Remember you are Bhagyashree, TL at HCL Tech.
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):

        """
        )

chain_email = prompt_email | llm
res = chain_email.invoke({"job_description": str(job), "link_list": links})
print(res.content)
```
